```python
from math import log10
def difference_of_magnitudes(brightness_ratio):
    difference_of_two_magnitudes = 2.5 * log10(brightness_ratio)
    return difference_of_two_magnitudes
brightness_ratio_star1_star2 = 100
result=difference_of_magnitudes(brightness_ratio_star1_star2)
print(f"The difference of magnitudes of the two stars- star1 and star2 is: {result}.")
```

```python
from math import log10
def difference_of_apparent_and_absolute_magnitudes(distance):
    distance_modulus = 5 * log10(distance / 10)  # distance is in parsecs
    return distance_modulus
distance_to_star = 100 #parsecs
result = difference_of_apparent_and_absolute_magnitudes(distance_to_star)
print(f"The difference of apparent and absolute magnitudes of the star is: {result}.")
```

```python
def wiens_Law(temperature):
    b = 2.898e-3 #m.K
    wavelength_max = b/temperature
    return wavelength_max
temperature_of_star = 5800 #K
result = wiens_Law(temperature_of_star)
print(f"The wavelength at which the star emits the most radiation is: {result} m.")
```

```python
from math import pi
def surface_area_of_sphere(radius):
    surface_area = 4*pi*radius**2
    return surface_area
radius_of_sphere = 4 #metre
result = surface_area_of_sphere(radius_of_sphere)
print(f"Surface area of the given sphere is: {result} m^2.")
```

```python
def leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    return False
current_year = 2026
confirmation_of_leap_year = leap_year(current_year)
print(f"The year {current_year} is a leap year: {confirmation_of_leap_year}.")
```
