```python
#creating class
class Telescope:
    def __init__(self, aperture, latitude, longitude):
        self.aperture = aperture
        self.latitude = latitude
        self.longitude = longitude
        #placeholder for target RA and Dec
        self.ra = None
        self.dec = None
    #defining method to pint out the telescope at given coordinates
    def point_at(self):
        target = input("Enter the RA, and Dec coordinates to point the telescope at (e.g. 12h30m00s +45d00m00s): ")
        coordinates = target.split()
        if len(coordinates) == 2:  
            self.ra = coordinates[0]
            self.dec = coordinates[1]
            print(f"The telescope target is fixed to RA: {self.ra} and Dec: {self.dec}")
        else:
            print("Invalid input. Please enter the RA and Dec coordinates in the correct format.")
    #defining method to take photo of the target
    def take_photo(self):
        if self.ra and self.dec:
            print(f"Taking a photo of the target at RA: {self.ra} and Dec: {self.dec} with an aperture of {self.aperture}mm.")
        else:
            print("No target coordinates set. Please point the telescope at a target before taking a photo.")
#creating an instance of the Telescope class
T1 = Telescope(200, 34.05, -118.25)
#pointing towards a target
    #target_coordinates = 12h30m00s +45d00m00s
T1.point_at()
#taking photo of the target
T1.take_photo()  
```

```python
import math
from datetime import datetime

def calculate_comet_distance(closest_approach_date_str: str, target_date_str: str, 
                             orbital_period_years: float, perihelion_distance_au: float) -> float:
    """
    Calculates the distance of a comet from the Sun (in AU) on a target date.
    
    Parameters:
    closest_approach_date_str (str): Date of perihelion in 'YYYY-MM-DD' format.
    target_date_str (str): Date to calculate distance for in 'YYYY-MM-DD' format.
    orbital_period_years (float): Orbital period of the comet in Earth years (T).
    perihelion_distance_au (float): Closest approach distance to the Sun in AU (q).
    
    Returns:
    float: Distance from the Sun in Astronomical Units (AU).
    """
    # 1. Parse dates and calculate time difference (t - T_0) in days
    date_format = "%Y-%m-%d"
    t0 = datetime.strptime(closest_approach_date_str, date_format)
    t = datetime.strptime(target_date_str, date_format)
    delta_days = (t - t0).days
    
    # 2. Derive Semi-major axis (a) using Kepler's Third Law (T^2 = a^3)
    # a = T^(2/3)
    a = orbital_period_years ** (2.0 / 3.0)
    
    # 3. Derive Eccentricity (e) from perihelion distance (q)
    # q = a * (1 - e) -> e = 1 - (q / a)
    e = 1.0 - (perihelion_distance_au / a)
    if e < 0 or e >= 1:
        raise ValueError("The derived orbit is not a valid ellipse (e must be between 0 and 1).")
        
    # 4. Calculate Mean Anomaly (M) in radians
    # M = n * t, where n = 2 * pi / Period_in_days
    period_days = orbital_period_years * 365.25
    mean_motion = (2.0 * math.pi) / period_days
    M = mean_motion * delta_days
    
    # Normalize M to be within [-pi, pi]
    M = (M + math.pi) % (2.0 * math.pi) - math.pi
    
    # 5. Solve Kepler's Equation (M = E - e * sin(E)) for Eccentric Anomaly (E) using Newton's method
    E = M  # Initial guess
    tolerance = 1e-10
    max_iterations = 100
    
    for _ in range(max_iterations):
        delta_E = (E - e * math.sin(E) - M) / (1.0 - e * math.cos(E))
        E -= delta_E
        if abs(delta_E) < tolerance:
            break
            
    # 6. Calculate Distance from Sun (r)
    # r = a * (1 - e * cos(E))
    r = a * (1.0 - e * math.cos(E))
    
    return r

# --- Example Usage (e.g., Comet Halley approximation) ---
# Closest approach (Perihelion): Feb 9, 1986
# Period: 75.3 years, Perihelion distance: 0.586 AU
dist = calculate_comet_distance("1986-02-09", "1988-02-09", 75.3, 0.586)
print(f"Calculated Distance from the Sun: {dist:.4f} AU")
```

```python
# Creating a Planet class
class Planet:
    def __init__(self, name, diameter):
        self.name = name
        self.diameter = diameter  # in kilometers

# Creating Solar System class
class Solar_System:
    def __init__(self, sun_name):
        self.sun_name = sun_name
        self.planets = []  # Solar system tracks its own planets
        self.planet_diameters = []  # Solar system tracks the diameters

    # Method belongs to the Solar_System class, not nested in __init__
    def add_planet(self, planet):
        self.planets.append(planet)
        self.planet_diameters.append(planet.diameter)

    # Method belongs to the Solar_System class, not nested in __init__
    def describe(self):
        print(f"The solar system has a star named {self.sun_name} and the following planets:")
        for planet in self.planets:
            print(f"- {planet.name} with a diameter of {planet.diameter} km")

# Creating an example of the solar system
solar_system = Solar_System("Sun")

# Creating planet instances
earth = Planet("Earth", 12742)
mars = Planet("Mars", 6779)
jupiter = Planet("Jupiter", 139820)

# Adding planets directly to the solar system instance
solar_system.add_planet(earth)
solar_system.add_planet(mars)
solar_system.add_planet(jupiter)

# Describing the solar system
solar_system.describe()
```

```python
class Galaxy:
    def __init__(self, name, galaxy_type, number_of_stars, presence_of_black_hole):
        self.name = name #galaxy name
        self.galaxy_type = galaxy_type #galaxy type
        self.number_of_stars = number_of_stars #no of stars
        self.presence_of_black_hole = presence_of_black_hole.lower() #presence of black hole in the galaxy, must be "yes", assuming every galaxy similar to MilyWay in size has a blachole at their center.
        if self.presence_of_black_hole != "yes":
            raise ValueError("Unable to construct the galaxy without a central black hole. Please provide valid input.")
#defining a method to calculate stellar mass of a galaxy
    def calculate_stellar_mass(self):
        stellar_mass = self.number_of_stars * 1.989e30
        return stellar_mass
#defining a method to simulate collision between two galaxies
    def collide_with(self, other_galaxy):
        #defining a condition to double check the presence of black hole in both galaxies before simulating collision
        if self.presence_of_black_hole == "yes" and other_galaxy.presence_of_black_hole == "yes":
            print(f"Simulation of collision between {self.name} and {other_galaxy.name} is starting...")
        else:
            print("One or more galaxies were not constructed properly. Collision cannot be simulated.")
#defining a method to calculate probability of collision between two galaxies
    def probability_of_collision(self, other_galaxy):
        probability = 0.5 #assuming a MW-M31 probability of collision of ~50% as per the research paper: Sawala, T., Delhomelle, J., Deason, A.J. et al. No certainty of a Milky Way–Andromeda collision. Nat Astron 9, 1206–1217 (2025). https://doi.org/10.1038/s41550-025-02563-1
        if probability >= 0.80:
            collision_result = "A head-on collision is highly likely."
        elif 0.40 <= probability < 0.80:
            collision_result = "The cosmic merger is uncertain; a near-miss or flyby is equally possible."
        else:
            collision_result = "The galaxies will likely miss each other entirely."
        return collision_result, probability
#defining a method to calculate time until collision between two galaxies
    def time_until_collision(self, other_galaxy):
        time = 4.5 #in billion years
        print(f"The estimated time until collision between {self.name} and {other_galaxy.name} is approximately {time} billion years.")
        return time
#defining a method to calculate the mass of the merged galaxy after collsion, if the probability of collision is above 40%; here we account only for the stellar mass of the merged galaxy, neglecting the dark matter and gas mass of the merged galaxy for simplicity
    def merged_galaxy_mass(self, other_galaxy):
        collision_result, probability = self.probability_of_collision(other_galaxy)
        if probability >= 0.40:
            merged_mass = self.calculate_stellar_mass() + other_galaxy.calculate_stellar_mass()
            print(f"The estimated mass of the merged galaxy after collision between {self.name} and {other_galaxy.name} is approximately {merged_mass:.2e} kg.")
            return None
        else:
            print("Since the galaxies are likely to miss each other, a merged galaxy will not form.")
            return None
#creating instance of the galaxy class
MilkyWay = Galaxy("Milky Way", "Spiral", 100000000000,"Yes")
Andromeda = Galaxy("Andromeda", "Spiral", 100000000000, "Yes")
#calculating stellar mass of the two galaxies
print(f"Stellar mass of {MilkyWay.name} is approximately {MilkyWay.calculate_stellar_mass():.2e} kg.")
print(f"Stellar mass of {Andromeda.name} is approximately {Andromeda.calculate_stellar_mass():.2e} kg.")
#simulating collision between the two galaxies
MilkyWay.collide_with(Andromeda)
#calculating the probability of collision between the two galaxies
collision_result, probability = MilkyWay.probability_of_collision(Andromeda)
print(f"Collision Result: {collision_result}")
print(f"Probability: {probability}")
#calculating the time until collision between the two galaxies
print(f"Time until collision: {MilkyWay.time_until_collision(Andromeda)} billion years")
#calculating the mass of the merged galaxy after collision, if the probability of collision is above 40%
MilkyWay.merged_galaxy_mass(Andromeda)
```

```python
class Pulsar:
    def __init__(self, name, period, magnetic_field_strength):
        self.name = name #name of the pulsar
        self.period = period #rotation period of the pulsar in seconds
        if self.period <=0:
            raise ValueError("Invalid period. The period must be a positive value.")
        self.magnetic_field_strength = magnetic_field_strength #magnetic field strength of the pulsar in Tesla
    #defining a method to calculat  the frequency of the pulsar
    def calculate_frequency(self):
        frequency = 1.0 / self.period
        return frequency
    #defining a method to calculate the EM emission of the pulsar using simple relationship: E ∝ B and E ∝ f; from which we can say: E ∝ Bf, hence, E = k*Bf, where k is the constant of proportionality
    def calculate_emission(self):
        k = 1e-10 #arbitrary constant of proportionality for demonstration purposes
        magnetic_field_strength = self.magnetic_field_strength
        frequency = self.calculate_frequency()
        emission = k * magnetic_field_strength * frequency
        return emission
#creating an instance of the Pulsar class
Crab_Pulsar = Pulsar("Crab Pulsar", 0.033, 1e8)
#calculating the frequency of the crab pulsar
print(f"The frequency of the {Crab_Pulsar.name} is approximately {Crab_Pulsar.calculate_frequency():.2f} Hz.")
#calculating the EM emission of the crab pulsar
print(f"The estimated EM emission of the {Crab_Pulsar.name} is approximately {Crab_Pulsar.calculate_emission():.2e} arbitrary units.")
```
